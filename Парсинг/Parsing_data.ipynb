```python
url_wikipedia = 'https://ru.wikipedia.org'
```

```python
url_MaleRus = 'https://ru.wikipedia.org/wiki/Категория:Русские_мужские_имена'
url_FemaleRus = 'https://ru.wikipedia.org/wiki/Категория:Русские_женские_имена'
url_MaleRus_q  = urllib.parse.quote(url_MaleRus, safe='/:')
url_FemaleRus_q  = urllib.parse.quote(url_FemaleRus, safe='/:')
print(url_MaleRus_q)
print(url_FemaleRus_q)
```

```python
all_MaleRus = []
all_FemaleRus = []
```

```python
MaleRus_bs = BeautifulSoup(urllib.request.urlopen(url_MaleRus_q),'lxml')
FemaleRus_bs = BeautifulSoup(urllib.request.urlopen(url_FemaleRus_q), 'lxml')
```

```python
for i in range(30): # Почему именно 30????
    # load links:
    bef = len(all_MaleRus)
    for links in MaleRus_bs.find_all('div', class_='mw-category-group'):
        for l in links.find_all('a'):
            all_MaleRus.append((l['title'], url_wikipedia+l['href']))
    aft = len(all_MaleRus)
    print('a-b: {}, ll: {}, tot: {}'.format(aft-bef, all_MaleRus[-1][0], aft))
    # find next page
    all_a = MaleRus_bs.find_all('a')
    MaleRus_bs = None
    for a in all_a:
        if a.text == 'Следующая страница':
            if a.has_attr('href'):
                next_page_ref = url_wikipedia + a['href']
                print('i: {}, npr: {}'.format(i, next_page_ref))
                MaleRus_bs = BeautifulSoup(urllib.request.urlopen(next_page_ref), 'lxml')
                break
    if MaleRus_bs is None:
        break
```
