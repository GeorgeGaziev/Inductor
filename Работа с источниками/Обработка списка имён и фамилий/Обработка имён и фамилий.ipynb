```python
import pandas as pd
import pickle
```

# Имена

```python
data=pd.read_csv("russian_names.csv",sep=";")
data.dropna(inplace=True)
```

```python
data.info()
```

```python
with open("D:/NamesFromGufo.pickle","rb") as f:
    tmpNames=pickle.load(f)
```

```python
russian_names={}
```

```python
all_names = [i for i in tmpNames]
```

```python
for i in data.values:
    if i[2]!=0 and i[0].lower() in all_names:
        russian_names[i[0].lower()]=[i[1],int(i[2])]
```

```python
russian_names
```

```python
#имена, не вошедшие в список частот
s=0
for j in all_names:
    if not(j in russian_names.keys()):
        s+=1
        print(j)
print (s)
```

```python
with open("names_extended.pickle","wb") as f:
    pickle.dump(russian_names,f)
```

# Фамилии

```python
data=pd.read_csv("russian_surnames.csv",sep=";")
data.dropna(inplace=True)
```

```python
data.info()
```

```python
russian_surnames={}
for i in data.values:
    if i[1]!=0:
        russian_surnames[i[0].lower()]=i[1]
```

```python
russian_surnames
```

```python
len(russian_surnames.values())
```

```python
with open("surnames_extended.pickle","wb") as f:
    pickle.dump(russian_surnames,f)
```
